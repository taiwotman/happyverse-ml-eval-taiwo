```python
# Step 2: Load Inputs

assets_path = project_root / "assets"
with open(assets_path / "sample_transcript.txt", "r") as f:
    transcript = f.read()

with open(assets_path / "job_description.txt", "r") as f:
    job_description = f.read()

with open(assets_path / "questions.json", "r") as f:
    questions = json.load(f)

```

```python
# Step 3: Run Evaluation
scores = evaluate_transcript(transcript, job_description, questions)

print("ðŸŽ¯ Evaluation Result")
print(json.dumps(scores, indent=2))

# Step 4: Generate Slide
create_slide(scores, out_path=assets_path / "slide_sample.pdf")

print("âœ… Slide generated at: assets/slide_sample.pdf")
```
